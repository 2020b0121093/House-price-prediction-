```python
# Step 1: Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Step 2: Load the Dataset
df = pd.read_csv('sample_house_prices.csv')  # make sure the file is in the same folder
```

```python
# Step 3: Quick Look at Data
print("First 5 rows:")
print(df.head())
print("\nShape of dataset:", df.shape)
```

```python
# Step 4: Check for Missing Values
print("\nMissing values:")
print(df.isnull().sum())
```

```python
# Step 5: Clean the Data (Drop rows with missing values for now)
df_clean = df.dropna()
```

```python
# Step 6: Summary Statistics
print("\nStatistical summary:")
print(df_clean.describe())
```

```python
# Step 7: Data Visualization
```

```python
# Histogram of Sale Price
plt.figure(figsize=(8,5))
sns.histplot(df_clean['SalePrice'], kde=True, color='green')
plt.title('Distribution of Sale Price')
plt.xlabel('Sale Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Step 8: Correlation Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df_clean.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Step 9: Scatter Plot - LotArea vs SalePrice
plt.figure(figsize=(8,5))
sns.scatterplot(x='LotArea', y='SalePrice', data=df_clean)
plt.title('Lot Area vs Sale Price')
plt.show()
```

```python
# Step 10: YearBuilt vs SalePrice
plt.figure(figsize=(8,5))
sns.boxplot(x='YearBuilt', y='SalePrice', data=df_clean)
plt.title('Year Built vs Sale Price')
plt.xticks(rotation=90)
plt.show()
```

```python
# Step 11: Save cleaned dataset
df_clean.to_csv('cleaned_house_prices.csv', index=False)
```
